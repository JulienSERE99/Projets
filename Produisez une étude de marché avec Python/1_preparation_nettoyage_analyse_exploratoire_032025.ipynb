#### Traitement des outliers

```python
from scipy.stats import shapiro

stat, p = shapiro(prix_prod_filtre["Indice Prix Production"])
print(f"Shapiro-Wilk test: Stat={stat:.3f}, p-value={p:.5f}")

if p > 0.05:
    print("Données normalement distribuées ✅")
else:
    print("Données NON normales ❌")
```

```python
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

outliers = detect_outliers_iqr(prix_prod_filtre, "Indice Prix Production")
print(outliers)
```

L'indice des prix à la production semble aberrant pour l'Argentine notamment. On lui impute la médiane.

```python
median_value = prix_prod_filtre["Indice Prix Production"].median()

# Remplacement de la valeur pour l'Argentine
prix_prod_filtre.loc[prix_prod_filtre["Zone"] == "Argentine", "Indice Prix Production"] = median_value
```
